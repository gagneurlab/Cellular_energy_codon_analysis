# Finding the major isoforms per tissue based on GTEx TPM file

```python
import pandas as pd
import numpy as np
```

```python
data_path = '../../data/'
```

```python
tpm_table_path = 'https://storage.googleapis.com/gtex_analysis_v7/rna_seq_data/GTEx_Analysis_2016-01-15_v7_RSEMv1.2.22_transcript_tpm.txt.gz'
```

```python
tpm_table = pd.read_csv(tpm_table_path, delimiter='\t')
```

```python
tpm_table
```

## Tissue Specific major transcript

```python
sample_ids = pd.read_csv('https://storage.googleapis.com/gtex_analysis_v7/annotations/GTEx_v7_Annotations_SampleAttributesDS.txt'
                        ,delimiter='\t')

tissue_sample_ids = sample_ids.loc[:, ['SAMPID', 'SMTS', 'SMTSD']]

tissue_sample_ids = tissue_sample_ids.set_index('SAMPID')
tissue_sample_ids
```

```python
len(sample_ids['SMTS'].unique())
```

```python
tpm_table = tpm_table.set_index(['gene_id', 'transcript_id'])

tpm_table = tpm_table.T
```

```python
tpm_table
```

```python
tissue_sample_ids.columns = pd.MultiIndex.from_product([tissue_sample_ids.columns, ['']]) #create dumb multiindex in order to merge with the table
tissue_sample_ids
```

```python
def compute_major_isoforms(table_chunk, tissue_sample_ids):
    
    # Merge sample info table with samples' TPM table
    tissue_table = table_chunk.join(tissue_sample_ids, how='inner')

    # Get the median of each transcript per tissue. Transpose dataframe so we can have multiindex with 
    # (gene_id, transcript) and columns (tissue_1, tissue_2,...tissue_n)
    tissue_table_median = tissue_table.groupby('SMTS').median().T

    # Get indexes of the transcript with highest median per gene
    major_isoforms = tissue_table_median.groupby('gene_id').apply(lambda group: group.idxmax())

    # Separate (gene_id, transcript_id) into transcript_id from the previous obtained indexes
    major_isoforms = major_isoforms.applymap(lambda element: element[1])

    return major_isoforms
```

Create table chunks idxs

```python
genes = tpm_table.columns.get_level_values(0).unique().values

gene_chunks = np.array_split(genes, 100)
```

Compute major isoforms for each chunk of genes:

```python
major_isoforms_chunks = []

n_chunks = len(gene_chunks)
for i in range(n_chunks):
    table_chunk = tpm_table.loc[:,gene_chunks[i]]
    major_isoforms_chunk = compute_major_isoforms(table_chunk, tissue_sample_ids)
    major_isoforms_chunks.append(major_isoforms_chunk)
    print(f'chunk {i+1} of {n_chunks} completed')

major_isoforms = pd.concat(major_isoforms_chunks)
```

```python
major_isoforms.to_csv(data_path+'gtex_major_isoform_per_major_tissue.csv')
```
